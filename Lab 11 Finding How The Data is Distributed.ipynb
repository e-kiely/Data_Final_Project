### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
# Display column names
print("Column Names:")
print(df.columns.tolist())

# Display data types for each column
print("\nData Types:")
print(df.dtypes)

# Display summary info of the dataset
print("\nSummary Information:")
df.info()
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
# 1. Identify missing values per column
missing_counts = df.isnull().sum()
print("Missing values per column:")
print(missing_counts[missing_counts > 0])

# 2. Impute missing values in categorical columns with the most frequent value
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        mode = df[col].mode()[0]
        df[col].fillna(mode, inplace=True)

# 3. Impute missing values in numerical columns with the median
numerical_cols = df.select_dtypes(include=['number']).columns
for col in numerical_cols:
    if df[col].isnull().sum() > 0:
        median = df[col].median()
        df[col].fillna(median, inplace=True)

# 4. Double-check if any missing values remain
print("\nMissing values after imputation:")
print(df.isnull().sum().sum())

# Optionally, if you prefer to drop rows with any missing values instead:
# df.dropna(inplace=True)
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
# List of key columns
key_columns = ['Employment', 'JobSat', 'YearsCodePro']

for col in key_columns:
    print(f"Value counts for '{col}':")
    print(df[col].value_counts(dropna=False))  # Include NaNs in the count
    print("\n" + "-"*40 + "\n")
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns

# Drop missing values for JobSat for cleaner visualization
job_sat_clean = df['JobSat'].dropna()

# Pie Chart for JobSat distribution
plt.figure(figsize=(8, 6))
job_sat_clean.value_counts().plot.pie(autopct='%1.1f%%', startangle=140, cmap='Pastel1')
plt.title('Job Satisfaction Distribution (Pie Chart)')
plt.ylabel('')
plt.show()

# KDE Plot for JobSat (assuming JobSat is numeric or ordinal)
plt.figure(figsize=(8, 6))
sns.kdeplot(job_sat_clean, shade=True)
plt.title('Job Satisfaction Distribution (KDE Plot)')
plt.xlabel('Job Satisfaction')
plt.ylabel('Density')
plt.show()

#Interpretation Example:
#The pie chart reveals the proportion of respondents across different satisfaction levels, showing which satisfaction scores are most common.

#The KDE plot smooths the distribution and helps identify trends such as peaks around specific satisfaction scores.

#If the distribution skews toward higher values, it indicates overall good job satisfaction; if it skews low or is bimodal, it suggests varied satisfaction levels among respondents.

#If you want me to tailor the interpretation based on your actual plot results, just share the insights or values!








```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Helper function to split the language lists
def split_languages(column):
    return df[column].dropna().str.split(';').explode()

# Get counts of languages worked with and want to work with
worked_counts = split_languages('LanguageHaveWorkedWith').value_counts()
want_counts = split_languages('LanguageWantToWorkWith').value_counts()

# Combine into one DataFrame for comparison (outer join to keep all languages)
lang_df = pd.DataFrame({
    'WorkedWith': worked_counts,
    'WantToWorkWith': want_counts
}).fillna(0).sort_values(by='WorkedWith', ascending=False)

# Select top N languages for clearer visualization
top_n = 10
lang_df_top = lang_df.head(top_n)

# Plot grouped bar chart
lang_df_top.plot(kind='bar', figsize=(12,6))
plt.title('Top Programming Languages: Worked With vs Want to Work With')
plt.xlabel('Programming Language')
plt.ylabel('Count of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example mapping of countries to regions (expand as needed)
country_to_region = {
    'United States': 'North America',
    'Canada': 'North America',
    'India': 'Asia',
    'Germany': 'Europe',
    'United Kingdom': 'Europe',
    'France': 'Europe',
    'Brazil': 'South America',
    'Australia': 'Oceania',
    # Add more countries/regions as needed
}

# Map countries to regions
df['Region'] = df['Country'].map(country_to_region)

# Filter out rows with unmapped countries or missing RemoteWork
filtered_df = df.dropna(subset=['Region', 'RemoteWork'])

# Create a count table of RemoteWork by Region
remote_region_counts = filtered_df.groupby(['Region', 'RemoteWork']).size().unstack(fill_value=0)

# Plot grouped bar chart
remote_region_counts.plot(kind='bar', figsize=(12, 7))
plt.title('Distribution of Remote Work by Region')
plt.xlabel('Region')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.legend(title='Remote Work')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here

import numpy as np

# Clean data by dropping missing values and converting YearsCodePro to numeric
clean_df = df[['JobSat', 'YearsCodePro']].dropna()
clean_df['YearsCodePro'] = pd.to_numeric(clean_df['YearsCodePro'], errors='coerce')
clean_df = clean_df.dropna()

# Extract the arrays
job_sat = clean_df['JobSat'].values
years_code_pro = clean_df['YearsCodePro'].values

# Calculate Pearson correlation manually using numpy
pearson_corr = np.corrcoef(job_sat, years_code_pro)[0,1]

print(f"Pearson correlation: {pearson_corr:.3f}")


```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Drop missing values in the two relevant columns
df_filtered = df[['Employment', 'EdLevel']].dropna()

# Create a cross-tabulation
crosstab = pd.crosstab(df_filtered['EdLevel'], df_filtered['Employment'])

# Plot a stacked bar chart
crosstab.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')

plt.title('Employment Type by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Employment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
