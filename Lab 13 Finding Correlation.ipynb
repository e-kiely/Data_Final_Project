### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns

# Drop missing or zero compensation values (optional: depends on your data)
comp_data = df['ConvertedCompYearly'].dropna()
comp_data = comp_data[comp_data > 0]

# Set plot style
sns.set(style="whitegrid")

# Create histogram and KDE distribution plot
plt.figure(figsize=(12, 6))
sns.histplot(comp_data, bins=50, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Yearly Compensation (ConvertedCompYearly)', fontsize=14)
plt.xlabel('Yearly Compensation (USD)', fontsize=12)
plt.ylabel('Number of Respondents', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
## Write your code here
# Filter for full-time employed respondents
full_time_df = df[df['Employment'] == 'Employed, full-time']

# Drop missing values in compensation column
comp_median = full_time_df['ConvertedCompYearly'].dropna().median()

print(f"Median compensation for full-time employed respondents: ${comp_median:,.2f}")
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns

# Filter dataset to relevant columns and remove rows with missing compensation or country
comp_by_country = df[['Country', 'ConvertedCompYearly']].dropna()

# Filter countries with sufficient responses (e.g., at least 50 respondents)
top_countries = comp_by_country['Country'].value_counts()
top_countries = top_countries[top_countries >= 50].index
filtered_df = comp_by_country[comp_by_country['Country'].isin(top_countries)]

# Plot boxplot
plt.figure(figsize=(14, 8))
sns.boxplot(x='ConvertedCompYearly', y='Country', data=filtered_df, showfliers=False)
plt.title('Compensation Distribution by Country (Excluding Outliers)', fontsize=14)
plt.xlabel('Converted Compensation (Yearly)')
plt.ylabel('Country')
plt.tight_layout()
plt.grid(True)
plt.show()

# --- Insights (as comments) ---
# - Boxplots reveal significant compensation variation across countries.
# - Countries like the United States and Switzerland typically show higher median compensations.
# - Some countries have narrow interquartile ranges, indicating consistent pay, while others (e.g., India) show broader spreads.
# - Extreme outliers are hidden for clarity but still exist, especially in high-income regions.
# - These insights can help identify where compensation disparities are most pronounced globally.
```
