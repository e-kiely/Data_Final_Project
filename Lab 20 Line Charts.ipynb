### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("survey-data.csv")

# Clean the Age column if necessary (strip spaces)
df['Age'] = df['Age'].str.strip()

# Group by Age and calculate median compensation, ignoring missing or zero values
median_comp = df[df['ConvertedCompYearly'] > 0].groupby('Age')['ConvertedCompYearly'].median()

# Sort Age groups logically if possible (assuming age groups like '18-24 years old', etc.)
age_order = [
    'Under 18 years old',
    '18-24 years old',
    '25-34 years old',
    '35-44 years old',
    '45-54 years old',
    '55-64 years old',
    '65 years or older',
    'Prefer not to say'
]
median_comp = median_comp.reindex(age_order)

# Plotting
plt.figure(figsize=(10,6))
plt.plot(median_comp.index, median_comp.values, marker='o', linestyle='-', color='b')
plt.title('Median Yearly Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Converted Compensation (Yearly)')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here
import matplotlib.pyplot as plt

# Define the age groups to include
age_subset = ['25-34 years old', '35-44 years old']

# Filter the data for these age groups and compensation > 0
df_filtered = df[(df['Age'].isin(age_subset)) & (df['ConvertedCompYearly'] > 0)]

# Calculate median compensation by age group
median_comp_subset = df_filtered.groupby('Age')['ConvertedCompYearly'].median()

# Order the age groups logically
median_comp_subset = median_comp_subset.reindex(age_subset)

# Plot
plt.figure(figsize=(8,5))
plt.plot(median_comp_subset.index, median_comp_subset.values, marker='o', linestyle='-', color='green')
plt.title('Median Yearly Compensation for Ages 25 to 45')
plt.xlabel('Age Group')
plt.ylabel('Median Converted Compensation (Yearly)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here
import numpy as np

def convert_years_code(x):
    if pd.isna(x):
        return np.nan
    elif x == "Less than 1 year":
        return 0.5
    elif x == "More than 50 years":
        return 51
    else:
        try:
            return int(x)
        except:
            return np.nan

df['YearsCodeProNum'] = df['YearsCodePro'].apply(convert_years_code)

# Now plot median JobSatPoints_6 by YearsCodeProNum
median_jobsat_by_exp = df.groupby('YearsCodeProNum')['JobSatPoints_6'].median().dropna()

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(median_jobsat_by_exp.index, median_jobsat_by_exp.values, marker='o')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Median Job Satisfaction (JobSatPoints_6)')
plt.title('Job Satisfaction by Years of Professional Coding Experience')
plt.grid(True)
plt.show()

```
